```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
#path to school file
school_data = Path("../Resources/schools_complete.csv")

#read school file
school_df = pd.read_csv(school_data)
school_df.head()
```

```python
#path to students file
student_data = Path("../Resources/students_complete.csv")

#read student file
student_df = pd.read_csv(student_data)
student_df.head()
```

```python
#Total number of unique schools, 
schools = school_df["school_name"].count()
schools #15

#Total students
total_students = student_df["student_name"].count()
total_students #39170

#Total budget

total_budget = school_df["budget"].sum()
total_budget #24649428

# average_math_score

average_math_score = student_df["maths_score"].mean()
average_math_score #70.33819249425581

#average_reading_score

average_reading_score = student_df["reading_score"].mean()
average_reading_score #69.98013786060761

#passing maths= total above 50/all *100

passing_math = ((student_df["maths_score"]>=50).sum())/\
             student_df["maths_score"].count()*100
passing_math   #86.07863160582077

# passing reading= total above 50/all *100
passing_reading = ((student_df["reading_score"]>=50).sum())/\
             student_df["reading_score"].count()*100
passing_reading  #84.42685728874139


# overall passing sum of students with 50 or above in math AND reading / total student number * 100

overall_passing = ([(student_df["maths_score"]>=50).sum() and ((student_df["reading_score"]>=50).sum())/\
                    (student_df["Student ID"].count())*100])

overall_passing [0]
```

```python
LGA_summary = pd.DataFrame({"Total School": [schools], 
                           "Total Students": [total_students], 
                           "Total Budget": [total_budget],
                           "Average Math Score": [average_math_score],
                           "Average Reading Score": [average_reading_score],
                           "% Passing Maths": [passing_math],
                           "% Passing Reading": [passing_reading],
                           "% Overall Passing": [overall_passing[0]]})
round(LGA_summary, 1)
```

```python
LGA_summary.set_index("Total School")
```

```python
merged_data = pd.merge(school_df, student_df, on=["school_name"])
merged_data.head()
```

```python
# school name

School_name = merged_data.groupby(["school_name"])
School_name.head()
```

```python
# school type

school_type = school_df.set_index('school_name')['type']
school_type
```

```python
total_students = School_name["School ID"].count()
total_students
```

```python
total_school_budget = school_df.set_index('school_name')['budget']
total_school_budget
```

```python
budget_per_student = (total_school_budget/total_students)
budget_per_student
```

```python
average_math_score = School_name["maths_score"].mean()
average_math_score
```

```python
average_reading_score = School_name["reading_score"].mean()
average_reading_score
```

```python
# % passing maths (the percentage of students who passed maths)
# (merged_data['maths_score'].groupby('school_name')['Student ID'].sum())*100 

pass_maths_number = merged_data[merged_data['maths_score'] >= 50].groupby('school_name')["Student ID"].count()

percent_pass_maths = (pass_maths_number/total_students)*100

percent_pass_maths
```

```python
# % passing reading (the percentage of students who passed reading)

pass_reading_number = merged_data[merged_data['reading_score'] >= 50].groupby('school_name')["Student ID"].count()

percent_pass_reading = (pass_reading_number/total_students)*100

percent_pass_reading
```

```python
#overall_passing = ([(student_df["maths_score"]>=50).sum() and ((student_df["reading_score"]>=50).sum())/\
                   #(student_df["Student ID"].count())*100])
    
    
percent_pass_overall = merged_data[(merged_data['maths_score'] >= 50) & (merged_data['reading_score'] >= 50)].groupby('school_name')['Student ID'].count()/\
                                    total_students*100
percent_pass_overall
```

```python
school_summary = pd.DataFrame({"School Type": school_type,
                               "Total Students": total_students, 
                               "Total School Budget": total_school_budget,
                               "Budget Per Student": budget_per_student,
                               "Average Maths Score": average_math_score,
                               "Average Reading Score": average_reading_score,
                               '% Passing Maths': percent_pass_maths,
                               '% Passing Reading': percent_pass_reading,
                               "% Overall Passing": percent_pass_overall})

school_summary.index.name = None


school_summary
```

```python
#Highest-Performing Schools (by % Overall Passing)
top_schools = school_summary.sort_values("% Overall Passing", ascending=False)
top_schools.head()
```

```python
#Sort the schools by % Overall Passing in ascending order and display the top 5 rows.
bottom_schools = school_summary.sort_values("% Overall Passing", ascending=True)
bottom_schools.head()
```

```python
#Scores by School Spending. Create a table that breaks down school performance based on average spending ranges (per student).
school_summary
```

```python
spending_bins = [0, 585, 630, 645, 680]

labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_summary["Spending Ranges (Per Student)"]=pd.cut(school_summary["Budget Per Student"], spending_bins, labels = labels)

school_summary
```

```python
spending_math_scores = school_summary.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = school_summary.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_summary.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
spending_summary = pd.DataFrame ({"Average Maths Score": spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Maths": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing": overall_passing_spending})
spending_summary
```

```python
round(spending_summary, 2)
```

```python
# Scores by School Size

size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary["School Size"]=pd.cut(school_summary["Total Students"], size_bins, labels = labels)

school_summary
```

```python
#Create a DataFrame called size_summary that breaks down school performance based on school size (small, medium, or large).

size_math_scores = school_summary.groupby(["School Size"])["Average Maths Score"].mean()
size_reading_scores = school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_summary.groupby(["School Size"])["% Passing Maths"].mean()
size_passing_reading = school_summary.groupby(["School Size"])["% Passing Reading"].mean()
overall_passing_size = school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
size_summary = pd.DataFrame ({"Average Maths Score": size_math_scores,
                                 "Average Reading Score": size_reading_scores,
                                 "% Passing Maths": size_passing_math,
                                 "% Passing Reading": size_passing_reading,
                                 "% Overall Passing": overall_passing_size})
size_summary
```

```python
#Scores by School Type. Use the per_school_summary DataFrame from the previous step to create a new DataFrame called type_summary.

type_math_scores = school_summary.groupby(["School Type"])["Average Maths Score"].mean()
type_reading_scores = school_summary.groupby(["School Type"])["Average Reading Score"].mean()
type_passing_math = school_summary.groupby(["School Type"])["% Passing Maths"].mean()
type_passing_reading = school_summary.groupby(["School Type"])["% Passing Reading"].mean()
overall_passing_type = school_summary.groupby(["School Type"])["% Overall Passing"].mean()
```

```python
type_summary = pd.DataFrame ({"Average Maths Score": type_math_scores,
                                 "Average Reading Score": type_reading_scores,
                                 "% Passing Maths": type_passing_math,
                                 "% Passing Reading": type_passing_reading,
                                 "% Overall Passing": overall_passing_type})
type_summary
```
